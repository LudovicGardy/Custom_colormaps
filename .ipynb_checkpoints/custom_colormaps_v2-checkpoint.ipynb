# Creating linearly variable color maps

The custom colormaps created here can vary from white-red to white-purple (color spectrum). 765 variations are currently possible, but it could be more or less by addind some small changes, if necessary.

Using the _**`create_custom_colormap function`**_, we get as an output _**`cmap`**_ and _**`color_map`**_. cmap is the matrix containing the rgb values. color_map is the object that can be used in matplotlib (imshow) as an actual colormap, on any image.

Let us define some functions.

```python
import math as Math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable, axes_size
import matplotlib.cm as cm

def get_all_cmaps(all_colormatrix):
    rgb_max_all_cmaps = []
    for i in range(all_colormatrix[1, :, 1].shape[0]):
        rgb_max_all_cmaps.append(all_colormatrix[1, i, :])
    all_colormaps = ListedColormap(rgb_max_all_cmaps)

    return(all_colormaps)


### Random add point
def getRandomInt(max):
    return Math.floor(Math.random() * max)


###  Random add point
def getRandomIntHex(max):
    return Math.floor(Math.random() * max).toString(16)


### Linear Interpolation
def lerp(inter, v0, v1):
    return Math.floor(inter*v0 + (1-inter)*v1)


### Distance Computation
def dist(p1, p2):
    return Math.sqrt((p2[0]-p1[0]) * (p2[0]-p1[0]) + (p2[1]-p1[1]) * (p2[1]-p1[1]))

def create_cmap(**kwargs):
    '''
    Create a color matrix and a color map using 3 lists of r (red), g (green) and b (blue) values.

    Parameters:
    - r (list of floats): red value, between 0 and 1
    - g (list of floats): green value, between 0 and 1
    - b (list of floats): blue value, between 0 and 1
    - mix_cmaps (boolean): for higher values un image enhancement
    - apply_segmentation (boolean): add a delineation between two ranges of values in the image (ex. cutoff: median)

    Returns:
    - color_matrix (numpy 2D array): contains all the rgb values for a given colormap
    - color_map (matplotlib object): the color_matrix transformed into an object that matplotlib can use on figures
    '''

    ### Init rgba matrix
    color_matrix = np.empty([256, 4])
    color_matrix.fill(0)

    ### Fill our empty matrix with our r,g,b values
    color_matrix[:, 0] = kwargs["r"]
    color_matrix[:, 1] = kwargs["g"]
    color_matrix[:, 2] = kwargs["b"]
    color_matrix[:, 3] = np.repeat(1, len(kwargs["b"]))  # alpha channel

    if kwargs["mix_cmaps"]:
        ### Mix 2 color maps: < our custom cmap > and < matplotlib RdGy cmap >
        RdGy_cmap = cm.get_cmap('gist_yarg', 256)
        RdGy_matrix = RdGy_cmap(np.linspace(0, 1, 256))

        data_selection1 = np.arange(0, 128, 1)
        data_selection2 = np.arange(0, 256, 2)
        cmap1_short = RdGy_matrix[data_selection1]
        cmap2_short = color_matrix[data_selection2]
        cmap_full = np.concatenate([cmap1_short, cmap2_short])
        
    if kwargs["mix_cmaps"] and kwargs["apply_segmentation"]:
        ### Add some transparency, except for the firt color of the color map (row=0)
        alpha = 0.6
        cmap_full[:, 3] = alpha
        cmap_full[0, 3] = 1
        
        ### Add a black delineation
        white = (1, 1, 1, 1)
        black = (0, 0, 0, 1)
        cmap_full[121:130,:] = black

        ### Convert our r,g,b matrix to a matplotlib colormap object
        #color_map = ListedColormap(color_matrix)

    if kwargs["mix_cmaps"]:        
        color_map = ListedColormap(cmap_full)
    else: 
        cmap_full = color_matrix
        color_map = ListedColormap(cmap_full)
        
    
    return(cmap_full, color_map) 

### Random image
def create_image():
    '''
    Create some random image on which we will apply the colormap. 
    Any other image could replace this one.
    '''

    dx, dy = 0.015, 0.05
    x = np.arange(-4.0, 4.0, dx)
    y = np.arange(-4.0, 4.0, dy)
    X, Y = np.meshgrid(x, y)

    def z_fun(x, y):
        return (1 - x / 2 + x**5 + y**6) * np.exp(-(x**2 + y**2))

    Z2 = z_fun(X, Y)
    return(Z2)
```

```python
def create_custom_colormap_variations(pos, colors, n_cmaps=100, cmap_length=256, power=5):
    '''
    Remake of Christophe Hurter's javascript implementation
    of Shepard's method on colors.
    '''

    ### Create table of colors
    #pos = [{"x": 0, "y": 0}, {"x": 0.25, "y": 0}, {"x": 0.5, "y": 0}, {"x": 0.75, "y": 0}, {"x": 1, "y": 0}]

    ##- Color set 1
    #colors = [{"r": 255, "g": 0, "b": 0}, {"r": 255, "g": 255, "b": 0}, {"r": 0, "g": 255, "b": 0}, {"r": 0, "g": 0, "b": 255}, {"r": 255, "g": 0, "b": 255}]

    ##- Color set 2
    #colors = [{"r": 0, "g": 0, "b": 255}, {"r": 10, "g": 230, "b": 220}, {"r": 0, "g": 255, "b": 0}, {"r": 230, "g": 220, "b": 10}, {"r": 255, "g": 0, "b": 0}]

    ### Create table of position
    n_colors = len(pos)  # len(pos)

    clientHeight = cmap_length  # n values in each color map
    clientWidth = n_cmaps  # n color maps

    all_colormatrix = []

    for y in range(0, clientHeight):
        all_colormatrix.append([])
        for x in range(0, clientWidth):
            ### Compute the distances
            p = (x/clientWidth, y/clientHeight)

            ### Compute the inverse distance weighting (IDW)
            d = list(np.repeat(0, len(pos)))

            colShepard = {"r": 0, "g": 0, "b": 0}

            for i in range(0, n_colors):
                d[i] = Math.pow(dist(p, [pos[i]["x"], pos[i]["y"]]), power)

            sumDist = 0
            for i in range(0, n_colors):
                if d[i] != 0:
                    sumDist += 1/d[i]

            ### Compute the interpolation
            for i in range(0, n_colors):
                if (d[i] == 0):
                    colShepard["r"] = colors[i]["r"]
                    colShepard["g"] = colors[i]["g"]
                    colShepard["b"] = colors[i]["b"]
                else:
                    colShepard["r"] += 1/d[i] * colors[i]["r"] / (sumDist)
                    colShepard["g"] += 1/d[i] * colors[i]["g"] / (sumDist)
                    colShepard["b"] += 1/d[i] * colors[i]["b"] / (sumDist)

            ### color hack toward white


            
            inter = {"r": 0, "g": 0, "b": 0}
            inter["r"] = colShepard["r"]
            inter["g"] = colShepard["g"]
            inter["b"] = colShepard["b"]
            
            if inter["r"] > 255:
                print(f"r: {inter['r']}")
                inter["r"] = 255
            if inter["g"] > 255:
                print(f"g: {inter['g']}")
                inter["g"] = 255
            if inter["b"] > 255:
                print(f"b: {inter['b']}")
                inter["b"] = 255

            #col = "#" + decimalToHex(colShepard["r"]) + decimalToHex(colShepard["g"]) + decimalToHex(colShepard["b"]) + "ff";
            col = [inter["r"]/255, inter["g"]/255, inter["b"]/255]
            all_colormatrix[-1].append(col)

    all_colormatrix = np.array(all_colormatrix)
    print(f"There are {all_colormatrix.shape[1]} color maps of size ({all_colormatrix.shape[0]},{all_colormatrix.shape[2]})")

    return(all_colormatrix)
```

```python
pos = [
       {"x": 0, "y": 0}, 
       {"x": 0.25, "y": 0}, 
       {"x": 0.5, "y": 0}, 
       {"x": 0.75, "y": 0}, 
       {"x": 1, "y": 0},
    
       {"x": 0, "y": 1}, 
       {"x": 0.25, "y": 1}, 
       {"x": 0.5, "y": 1}, 
       {"x": 0.75, "y": 1}, 
       {"x": 1, "y": 1}    
      ]

colors = [
       {"r": 0, "g": 0, "b": 255}, 
       {"r": 10, "g": 230, "b": 220}, 
       {"r": 0, "g": 255, "b": 0}, 
       {"r": 230, "g": 220, "b": 10}, 
       {"r": 255, "g": 0, "b": 0},
    
       {"r": 255, "g": 255, "b": 255}, 
       {"r": 255, "g": 255, "b": 255}, 
       {"r": 255, "g": 255, "b": 255}, 
       {"r": 255, "g": 255, "b": 255}, 
       {"r": 255, "g": 255, "b": 255}    
]

all_colormatrix = create_custom_colormap_variations(pos, colors, n_cmaps=1000, cmap_length=256)
all_colormaps = get_all_cmaps(all_colormatrix)
```

# Apply linear variation of colormaps regarding a weight (w) value

```python
### Let us create some image (any other could be used).
Z2 = create_image()

### Now create a color map, using the w value you want 0 = white-red, 765 = white-purple.
f, ax = plt.subplots(4,4, figsize= (20,20))

col = 0
for i, w in enumerate(np.arange(0,all_colormatrix.shape[1],int(all_colormatrix.shape[1]/15))):
    
    row = int(i/4)
    if row > 3:
        break
        
    #- Create image
    cmap, color_map = create_cmap(r=all_colormatrix[:,w,:][::-1][:,0],
                                  g=all_colormatrix[:,w,:][::-1][:,1],
                                  b=all_colormatrix[:,w,:][::-1][:,2],
                                  mix_cmaps=False,
                                  apply_segmentation=False)
    
    img = ax[row,col].imshow(Z2, cmap =color_map, alpha=0.7, interpolation ='bilinear',aspect = "auto")
    ax[row,col].set_title("w = {} | Pos. ({},{})".format(w,row,col))    
    ax[row,col].set_yticks([])
    ax[row,col].set_xticks([])
    
    #- Create colorbar
    aspect = 20
    pad_fraction = 0.5
    divider = make_axes_locatable(ax[row,col])
    height = axes_size.AxesX(ax[row,col], aspect=2/100)
    pad = axes_size.Fraction(0.5, height)
    cax = divider.append_axes("right", size=height, pad=pad)
    cax.tick_params(axis='both', which='major')
    plt.colorbar(img, cax=cax, orientation="vertical")    
    cax.set_yticks([])
    cax.set_xticks([]) 

    col +=1
    if col == 4:
        col = 0  
        

    
```

# Add a delineation/segmentation on the image

Here we fix all the pixels for which the value == 1 to black. So the "inactive" parts of the image are separated from the "active" parts.

```python
### Let us create some image (any other could be used).
Z2 = create_image()

### Now create a color map, using the w value you want 0 = white-red, 765 = white-purple.
f, ax = plt.subplots(4,4, figsize= (20,20))

col = 0
for i, w in enumerate(np.arange(0,all_colormatrix.shape[1],int(all_colormatrix.shape[1]/15))):
    
    row = int(i/4)
    if row > 3:
        break
        

    #- Create image
    color_map = ListedColormap(all_colormatrix[:,w,:][::-1])
    cmap, color_map = create_cmap(r=all_colormatrix[:,w,:][::-1][:,0],
                                  g=all_colormatrix[:,w,:][::-1][:,1],
                                  b=all_colormatrix[:,w,:][::-1][:,2],
                                  mix_cmaps=True,
                                  apply_segmentation=True)
    
    img = ax[row,col].imshow(Z2, cmap =color_map, alpha=0.7, interpolation ='bilinear',aspect = "auto")
    ax[row,col].set_title("w = {} | Pos. ({},{})".format(w,row,col))    
    ax[row,col].set_yticks([])
    ax[row,col].set_xticks([]) 
    
    #- Create colorbar
    aspect = 20
    pad_fraction = 0.5
    divider = make_axes_locatable(ax[row,col])
    height = axes_size.AxesX(ax[row,col], aspect=2/100)
    pad = axes_size.Fraction(0.5, height)
    cax = divider.append_axes("right", size=height, pad=pad)
    cax.tick_params(axis='both', which='major')
    plt.colorbar(img, cax=cax, orientation="vertical")    
    cax.set_yticks([])
    cax.set_xticks([])   
    
    col +=1
    if col == 4:
        col = 0        
    
```

```python
def create_custom_colromap_1D(cmap_length=100):

    clientWidth = cmap_length

    ### Gradient 1D

    ###Defines the first color
    color1 = {"r": 200, "g": 200, "b": 200}
    pos1 = {"x": 0, "y": 0}

    ###Defines the second color
    color2 = {"r": 0, "g": 0, "b": 0}

    ###Defines the third color
    color3 = {"r": 255, "g": 255, "b": 255}

    ###Defines the 4rd color
    color4 = {"r": 0, "g": 0, "b": 255}

    colormatrix = []
    colormatrix.append([200,200,200,1])
    colormatrix.append([0,0,0,1])
    
    for x in range(0, clientWidth):
        ###Compute the interpolation
        t = x/clientWidth
        colormatrix.append([lerp(t, color2["r"], color1["r"])/255,
                            lerp(t, color2["g"], color1["g"])/255,
                            lerp(t, color2["b"], color1["b"])/255,
                            1])

    for x in range(0, clientWidth):
        ###Compute the interpolation
        t = x/clientWidth
        colormatrix.append([lerp(t, color3["r"], color2["r"])/255,
                            lerp(t, color3["g"], color2["g"])/255,
                            lerp(t, color3["b"], color2["b"])/255,
                            1])


    colormatrix = np.array(colormatrix)
    return(colormatrix)
```
